```python
import xarray as xr
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from tqdm import tqdm
import os

# ============================================================
# 1. DATA LOADING & PREPROCESSING
# ============================================================

rain_01_path = "rainfall_2020_01degree_combined.nc"
ds = xr.open_dataset(rain_01_path)

# Extract India region
rain = ds["precipitation"].sel(
    lat=slice(38, 6),
    lon=slice(68, 98)
)

# Store lat/lon for visualization
lats = rain.lat.values
lons = rain.lon.values

# Convert to tensor (T, H, W)
rain_tensor = torch.FloatTensor(rain.values)

# Add channel → (T, 1, H, W)
rain_tensor = rain_tensor.unsqueeze(1)

# Downsample (0.1° → 0.5°)  ← kept separate for plotting
low_res = F.avg_pool2d(rain_tensor, kernel_size=5, stride=5)

# Upsample back to original size (bilinear baseline fed to model)
low_res_up = F.interpolate(low_res, scale_factor=5, mode="bilinear", align_corners=False)

# Lat/lon grids at 0.5° resolution (for the extra panel)
lats_05 = lats[::5]
lons_05 = lons[::5]

print(f"High-res shape  : {rain_tensor.shape}")
print(f"Low-res 0.5° shape: {low_res.shape}")
print(f"Upsampled shape : {low_res_up.shape}")

# Log transform (stabilises rainfall distribution)
X = torch.log1p(low_res_up)
Y = torch.log1p(rain_tensor)

# ============================================================
# 2. TRAIN / VAL SPLIT
# ============================================================

T = X.shape[0]
split = int(0.85 * T)

X_train, X_val = X[:split], X[split:]
Y_train, Y_val = Y[:split], Y[split:]

# Keep the raw low-res val slice for plotting
low_res_val = low_res[split:]

print(f"Train samples: {X_train.shape[0]}  |  Val samples: {X_val.shape[0]}")

# ============================================================
# 3. DATASET & DATALOADER
# ============================================================

class RainDataset(Dataset):
    def __init__(self, X, Y):
        self.X = X
        self.Y = Y

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.X[idx], self.Y[idx]

train_dataset = RainDataset(X_train, Y_train)
val_dataset   = RainDataset(X_val,   Y_val)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True,  num_workers=8, pin_memory=True)
val_loader   = DataLoader(val_dataset,   batch_size=32, shuffle=False, num_workers=8, pin_memory=True)

# ============================================================
# 4. MODEL DEFINITION
# ============================================================

class DownscalingUNet(nn.Module):
    def __init__(self):
        super(DownscalingUNet, self).__init__()

        # ---- Encoder ----
        self.c1_1     = nn.Conv2d(1,  64, kernel_size=3, padding=1)
        self.c1_2     = nn.Conv2d(64, 32, kernel_size=3, padding=1)
        self.bn1_1    = nn.BatchNorm2d(64)
        self.bn1_2    = nn.BatchNorm2d(32)
        self.dropout1 = nn.Dropout(0.3)

        self.c2_1     = nn.Conv2d(32, 32, kernel_size=3, padding=1)
        self.c2_2     = nn.Conv2d(32, 32, kernel_size=3, padding=1)
        self.c2_3     = nn.Conv2d(32, 16, kernel_size=3, padding=1)
        self.bn2_1    = nn.BatchNorm2d(32)
        self.bn2_2    = nn.BatchNorm2d(32)
        self.bn2_3    = nn.BatchNorm2d(16)
        self.dropout2 = nn.Dropout(0.2)

        self.c3_1     = nn.Conv2d(16, 16, kernel_size=3, padding=1)
        self.c3_2     = nn.Conv2d(16, 16, kernel_size=3, padding=1)
        self.c3_3     = nn.Conv2d(16, 16, kernel_size=3, padding=1)
        self.bn3_1    = nn.BatchNorm2d(16)
        self.bn3_2    = nn.BatchNorm2d(16)
        self.bn3_3    = nn.BatchNorm2d(16)
        self.dropout3 = nn.Dropout(0.1)

        self.maxpool = nn.MaxPool2d(2)
        self.avgpool = nn.AvgPool2d(2)

        # ---- Decoder ----
        self.up4      = nn.ConvTranspose2d(16, 32, kernel_size=2, stride=2)
        self.c4_1     = nn.Conv2d(48, 32, kernel_size=3, padding=1)
        self.c4_2     = nn.Conv2d(32, 32, kernel_size=3, padding=1)
        self.bn4_1    = nn.BatchNorm2d(32)
        self.bn4_2    = nn.BatchNorm2d(32)
        self.dropout4 = nn.Dropout(0.2)

        self.up5      = nn.ConvTranspose2d(32, 16, kernel_size=2, stride=2)
        self.c5_1     = nn.Conv2d(48, 16, kernel_size=3, padding=1)
        self.c5_2     = nn.Conv2d(16, 16, kernel_size=3, padding=1)
        self.bn5_1    = nn.BatchNorm2d(16)
        self.bn5_2    = nn.BatchNorm2d(16)
        self.dropout5 = nn.Dropout(0.1)

        self.out = nn.Conv2d(16, 1, kernel_size=1)

    def forward(self, x):
        c1 = F.relu(self.bn1_1(self.c1_1(x)))
        c1 = self.dropout1(c1)
        c1 = F.relu(self.bn1_2(self.c1_2(c1)))

        p1 = self.maxpool(c1)
        a1 = self.avgpool(c1)

        c2 = F.relu(self.bn2_1(self.c2_1(p1)))
        c2 = F.relu(self.bn2_2(self.c2_2((p1 + c2) * 0.5)))
        c2 = self.dropout2(c2)
        c2 = F.relu(self.bn2_3(self.c2_3((c2 + a1) * 0.5)))

        p2 = self.maxpool(c2)
        a2 = self.avgpool(c2)

        c3 = F.relu(self.bn3_1(self.c3_1(p2)))
        c3 = F.relu(self.bn3_2(self.c3_2((p2 + c3) * 0.5)))
        c3 = self.dropout3(c3)
        c3 = F.relu(self.bn3_3(self.c3_3((c3 + a2) * 0.5)))

        u4 = self.up4(c3)
        u4 = (u4 + p1) * 0.5
        u4 = torch.cat([u4, c2], dim=1)
        c4 = F.relu(self.bn4_1(self.c4_1(u4)))
        c4 = self.dropout4(c4)
        c4 = F.relu(self.bn4_2(self.c4_2(c4)))

        u5 = self.up5(c4)
        u5 = torch.cat([u5, c1], dim=1)
        c5 = F.relu(self.bn5_1(self.c5_1(u5)))
        c5 = self.dropout5(c5)
        c5 = F.relu(self.bn5_2(self.c5_2(c5)))

        return F.relu(self.out(c5))

# ============================================================
# 5. COMBINED LOSS: MAE + GRADIENT LOSS
# ============================================================

def gradient_loss(pred, target):
    def grad(x):
        dy = x[:, :, 1:, :] - x[:, :, :-1, :]
        dx = x[:, :, :, 1:] - x[:, :, :, :-1]
        return dy, dx
    dy_p, dx_p = grad(pred)
    dy_t, dx_t = grad(target)
    return F.l1_loss(dy_p, dy_t) + F.l1_loss(dx_p, dx_t)

def combined_loss(pred, target, alpha=0.8):
    return alpha * F.l1_loss(pred, target) + (1 - alpha) * gradient_loss(pred, target)

# ============================================================
# 6. TRAINING LOOP
# ============================================================

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

model     = DownscalingUNet().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode="min", patience=5, factor=0.5
)

EPOCHS = 100
best_val = float("inf")
train_losses, val_losses = [], []

for epoch in range(1, EPOCHS + 1):
    model.train()
    train_loss = 0.0
    for xb, yb in tqdm(train_loader, desc=f"Epoch {epoch:02d}/{EPOCHS} [Train]", leave=False):
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        pred = model(xb)
        loss = combined_loss(pred, yb)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        train_loss += loss.item() * xb.size(0)
    train_loss /= len(train_dataset)

    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            pred = model(xb)
            val_loss += combined_loss(pred, yb).item() * xb.size(0)
    val_loss /= len(val_dataset)

    train_losses.append(train_loss)
    val_losses.append(val_loss)
    scheduler.step(val_loss)

    print(f"Epoch {epoch:02d}/{EPOCHS} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

    if val_loss < best_val:
        best_val = val_loss
        torch.save(model.state_dict(), "best_downscaling_model.pth")
        print(f"  ✓ Best model saved (val_loss={best_val:.4f})")

print("\nTraining complete!")

# ============================================================
# 7. LOAD BEST MODEL
# ============================================================

model.load_state_dict(torch.load("best_downscaling_model.pth", map_location=device))
model.eval()

# ============================================================
# 8. METRICS  (RMSE, MAE, Bias, R², R² rainy-only)
# ============================================================

def compute_metrics(model, loader, device):
    all_pred, all_true = [], []
    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device)
            pred = model(xb).cpu()
            all_pred.append(torch.expm1(pred))
            all_true.append(torch.expm1(yb))

    pred = torch.cat(all_pred).flatten()
    true = torch.cat(all_true).flatten()

    rmse = torch.sqrt(F.mse_loss(pred, true)).item()
    mae  = F.l1_loss(pred, true).item()
    bias = (pred - true).mean().item()

    # Overall R²
    ss_res = ((true - pred) ** 2).sum()
    ss_tot = ((true - true.mean()) ** 2).sum()
    r2 = (1 - ss_res / ss_tot).item()

    # R² on rainy pixels only (> 1 mm/day)
    mask = true > 1.0
    pred_r, true_r = pred[mask], true[mask]
    ss_res_r = ((true_r - pred_r) ** 2).sum()
    ss_tot_r = ((true_r - true_r.mean()) ** 2).sum()
    r2_rainy = (1 - ss_res_r / ss_tot_r).item() if ss_tot_r > 0 else float("nan")

    return rmse, mae, bias, r2, r2_rainy

rmse, mae, bias, r2, r2_rainy = compute_metrics(model, val_loader, device)

print(f"\n=== Validation Metrics (original scale) ===")
print(f"  RMSE            : {rmse:.4f} mm")
print(f"  MAE             : {mae:.4f} mm")
print(f"  Bias            : {bias:.4f} mm")
print(f"  R²  (all pixels): {r2:.4f}")
print(f"  R²  (>1 mm/day) : {r2_rainy:.4f}")

# ============================================================
# 9. VISUALIZATION
# ============================================================

rain_colors = [
    "#ffffff", "#b3d9ff", "#66b3ff", "#0080ff",
    "#00cc44", "#ffff00", "#ff8800", "#ff0000", "#800000"
]
rain_cmap = mcolors.LinearSegmentedColormap.from_list("rainfall", rain_colors, N=256)

def to_mm(t):
    """Inverse log-transform → mm/day numpy (H, W)."""
    return torch.expm1(t).squeeze().cpu().numpy()

# Sample day from validation set
sample_idx = 0
x_sample = X_val[sample_idx].unsqueeze(0).to(device)
y_sample  = Y_val[sample_idx]

with torch.no_grad():
    pred_sample = model(x_sample).cpu().squeeze(0)

# ── Four fields ──────────────────────────────────────────────
lr_05_mm = low_res_val[sample_idx].squeeze().numpy()          # raw 0.5° (mm)
lr_mm    = to_mm(X_val[sample_idx])                           # bilinear upsampled input
pred_mm  = to_mm(pred_sample)                                  # model prediction
hr_mm    = to_mm(y_sample)                                     # ground truth 0.1°

vmax = max(lr_05_mm.max(), lr_mm.max(), pred_mm.max(), hr_mm.max(), 1.0)

# ── Figure 1: Four-panel comparison ──────────────────────────
fig, axes = plt.subplots(1, 4, figsize=(24, 7),
                         subplot_kw={"projection": ccrs.PlateCarree()},
                         facecolor="#0d1117")

panels = [
    (lr_05_mm, lats_05, lons_05, "Raw 0.5° Input\n(Coarse Grid)",      "#ce93d8"),
    (lr_mm,    lats,    lons,    "Bilinear Upsampled\n(0.5° → 0.1°)",  "#4fc3f7"),
    (pred_mm,  lats,    lons,    "UNet Prediction\n(0.1°)",            "#ffd54f"),
    (hr_mm,    lats,    lons,    "Ground Truth\n(0.1°)",               "#a5d6a7"),
]

for ax, (field, lat_g, lon_g, title, col) in zip(axes, panels):
    ax.set_facecolor("#0d1117")
    ax.set_extent([68, 98, 6, 38], crs=ccrs.PlateCarree())

    im = ax.pcolormesh(lon_g, lat_g, field,
                       cmap=rain_cmap, vmin=0, vmax=vmax,
                       transform=ccrs.PlateCarree())

    ax.add_feature(cfeature.COASTLINE.with_scale("50m"), linewidth=0.8, edgecolor="#e0e0e0")
    ax.add_feature(cfeature.BORDERS.with_scale("50m"),   linewidth=0.6, edgecolor="#aaaaaa")
    ax.add_feature(cfeature.STATES.with_scale("50m"),    linewidth=0.3, edgecolor="#666666")

    gl = ax.gridlines(draw_labels=True, linewidth=0.3,
                      color="#444444", alpha=0.7, linestyle="--")
    gl.top_labels = gl.right_labels = False
    gl.xlabel_style = gl.ylabel_style = {"color": "#cccccc", "fontsize": 7}

    ax.set_title(title, color=col, fontsize=12, fontweight="bold", pad=10)

    cb = plt.colorbar(im, ax=ax, orientation="horizontal",
                      pad=0.04, fraction=0.04, aspect=30)
    cb.set_label("Rainfall (mm/day)", color="#cccccc", fontsize=8)
    cb.ax.tick_params(colors="#cccccc", labelsize=7)

fig.suptitle("Rainfall Downscaling over India — 0.5° → 0.1° UNet",
             color="#ffffff", fontsize=16, fontweight="bold", y=1.01)
plt.tight_layout()
plt.savefig("india_rainfall_comparison.png", dpi=150,
            bbox_inches="tight", facecolor="#0d1117")
plt.show()
print("Saved: india_rainfall_comparison.png")

# ── Figure 2: Error maps ──────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 6),
                         subplot_kw={"projection": ccrs.PlateCarree()},
                         facecolor="#0d1117")

error     = pred_mm - hr_mm
abs_error = np.abs(error)
elim      = np.percentile(abs_error, 98)

for ax, (field, cmap, title, col, label) in zip(axes, [
    (error,     "RdBu_r", "Signed Error (Pred − Truth)", "#ef9a9a", "mm/day"),
    (abs_error, "hot_r",  "Absolute Error |Pred − Truth|", "#ffcc80", "mm/day"),
]):
    ax.set_facecolor("#0d1117")
    ax.set_extent([68, 98, 6, 38], crs=ccrs.PlateCarree())
    kwargs = dict(vmin=-elim, vmax=elim) if "Signed" in title else dict(vmin=0, vmax=elim)
    im = ax.pcolormesh(lons, lats, field, cmap=cmap,
                       transform=ccrs.PlateCarree(), **kwargs)
    ax.add_feature(cfeature.COASTLINE.with_scale("50m"), linewidth=0.8, edgecolor="#e0e0e0")
    ax.add_feature(cfeature.BORDERS.with_scale("50m"),   linewidth=0.6, edgecolor="#aaaaaa")
    ax.add_feature(cfeature.STATES.with_scale("50m"),    linewidth=0.3, edgecolor="#666666")
    gl = ax.gridlines(draw_labels=True, linewidth=0.3, color="#444444", linestyle="--")
    gl.top_labels = gl.right_labels = False
    gl.xlabel_style = gl.ylabel_style = {"color": "#cccccc", "fontsize": 8}
    ax.set_title(title, color=col, fontsize=13, fontweight="bold")
    cb = plt.colorbar(im, ax=ax, orientation="horizontal", pad=0.04, fraction=0.04)
    cb.set_label(label, color="#cccccc", fontsize=9)
    cb.ax.tick_params(colors="#cccccc", labelsize=8)

fig.suptitle("Prediction Error Maps over India", color="#ffffff",
             fontsize=15, fontweight="bold", y=1.01)
plt.tight_layout()
plt.savefig("india_rainfall_error.png", dpi=150,
            bbox_inches="tight", facecolor="#0d1117")
plt.show()
print("Saved: india_rainfall_error.png")

# ── Figure 3: Training curves ─────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 5), facecolor="#0d1117")
ax.set_facecolor("#161b22")

ep = range(1, len(train_losses) + 1)
ax.plot(ep, train_losses, color="#4fc3f7", linewidth=2, label="Train Loss")
ax.plot(ep, val_losses,   color="#ffd54f", linewidth=2, linestyle="--", label="Val Loss")
ax.fill_between(ep, train_losses, alpha=0.15, color="#4fc3f7")
ax.fill_between(ep, val_losses,   alpha=0.15, color="#ffd54f")

ax.set_xlabel("Epoch", color="#cccccc", fontsize=12)
ax.set_ylabel("Combined Loss (MAE + Gradient)", color="#cccccc", fontsize=12)
ax.set_title("Training & Validation Loss Curves", color="#ffffff",
             fontsize=14, fontweight="bold")
ax.legend(facecolor="#0d1117", edgecolor="#444444", labelcolor="#cccccc", fontsize=11)
ax.tick_params(colors="#cccccc")
for spine in ax.spines.values():
    spine.set_edgecolor("#333333")
ax.grid(True, color="#333333", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.savefig("training_curves.png", dpi=150, bbox_inches="tight", facecolor="#0d1117")
plt.show()
print("Saved: training_curves.png")

# ── Figure 4: Scatter — predicted vs truth with R² ───────────
flat_pred = pred_mm.flatten()
flat_true = hr_mm.flatten()

rng  = np.random.default_rng(42)
mask = rng.choice(len(flat_pred), size=min(5000, len(flat_pred)), replace=False)

fig, ax = plt.subplots(figsize=(7, 7), facecolor="#0d1117")
ax.set_facecolor("#161b22")

sc = ax.scatter(flat_true[mask], flat_pred[mask], c=flat_true[mask],
                cmap=rain_cmap, alpha=0.5, s=8, vmin=0, vmax=vmax)
ax.plot([0, vmax], [0, vmax], color="#ff4444", linewidth=1.5,
        linestyle="--", label="1:1 line")

cb = plt.colorbar(sc, ax=ax, fraction=0.04)
cb.set_label("True Rainfall (mm/day)", color="#cccccc", fontsize=10)
cb.ax.tick_params(colors="#cccccc")

ax.set_xlabel("Ground Truth (mm/day)", color="#cccccc", fontsize=12)
ax.set_ylabel("Predicted (mm/day)",    color="#cccccc", fontsize=12)
ax.set_title("Predicted vs Ground Truth — Pixel-Level", color="#ffffff",
             fontsize=13, fontweight="bold")
ax.legend(facecolor="#0d1117", edgecolor="#444444", labelcolor="#cccccc")
ax.tick_params(colors="#cccccc")
for spine in ax.spines.values():
    spine.set_edgecolor("#333333")
ax.grid(True, color="#333333", linestyle="--", alpha=0.5)

# ── Metrics box now includes both R² values ──
metrics_text = (
    f"RMSE = {rmse:.3f} mm\n"
    f"MAE  = {mae:.3f} mm\n"
    f"Bias = {bias:.3f} mm\n"
    f"R²   = {r2:.4f}  (all pixels)\n"
    f"R²   = {r2_rainy:.4f}  (>1 mm/day)"
)
ax.text(0.05, 0.87, metrics_text, transform=ax.transAxes,
        color="#ffd54f", fontsize=9.5,
        bbox=dict(facecolor="#0d1117", edgecolor="#444444",
                  alpha=0.85, boxstyle="round"))

plt.tight_layout()
plt.savefig("scatter_pred_vs_truth.png", dpi=150,
            bbox_inches="tight", facecolor="#0d1117")
plt.show()
print("Saved: scatter_pred_vs_truth.png")

print("\n✅ All done! Outputs:")
print("   • best_downscaling_model.pth")
print("   • india_rainfall_comparison.png   ← now 4 panels (0.5°, bilinear, UNet, truth)")
print("   • india_rainfall_error.png")
print("   • training_curves.png")
print("   • scatter_pred_vs_truth.png       ← R² (all) + R² (rainy) annotated")
```
